```python
import math
import json
import sys
import pandas as pd
import sys
import numpy as np
from PIL import Image
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
cutFilesFolder='data/bihar/bihar_2010_landsat7_cutFiles_rgb'
villageCutFilesList = [f for f in listdir(cutFilesFolder) if isfile(join(cutFilesFolder, f))]
```

```python
listFlattenArray=[None,None,None]
count_v=0
for cxyz in villageCutFilesList:
    count_v+=1
    if(count_v % 1000 == 0):
        print(count_v)
    if(count_v > 15000):
        break
    #currVillageFile=cutFilesFolder+'/'+villageCutFilesList[0]
    currVillageFile=cutFilesFolder+'/'+cxyz
    img = Image.open( currVillageFile )
    img.load()
    data = np.asarray( img )
    for tz in range(3):
        flattenData=data[:,:,tz].flatten()
        flattenData=flattenData[flattenData != 0]
        if(listFlattenArray[tz] is None):
            listFlattenArray[tz] = flattenData
        else:
            listFlattenArray[tz] = np.append(listFlattenArray[tz],flattenData)
```

```python
x=listFlattenArray[0]
bins = np.linspace(min(x),max(x), 20)
plt.clf()
plt.title('Red',fontsize=30)
plt.xlabel('Random Histogram')
plt.ylabel('Frequency',fontsize=30)
plt.hist(x, bins, alpha=0.5, histtype='bar', ec='black')
plt.legend(loc='upper right',fontsize=30)
plt.xticks(fontsize = 20) 
plt.yticks(fontsize = 20) 
plt.show()
```
